```python
import pandas as pd
import sklearn.neighbors as n
import numpy as np
```

```python
df = pd.read_csv("data/breast-cancer/data.csv")
df
```

```python
y = df["diagnosis"]
x = df.drop(["diagnosis", "id"], axis=1)
```

```python
model = n.KNeighborsClassifier(n_neighbors=3)
model.fit(x, y)
model.score(x, y)
```
