```python
df = df.drop(["slope", "ca", "thal", "fbs"], axis=1)
df
```

```python
mean_chol = np.mean(df["chol"])
std_chol = np.std(df["chol"])
mean_chol, std_chol
```

```python
df["chol"] = df["chol"].fillna(mean_chol + (np.random.rand() - 0.5) * std_chol)
```

```python
df = df.dropna()
df.to_csv("data/heartdisease/dataclean.csv")
```
