```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a1 = np.arange(-2 * np.pi, 2 * np.pi, 0.01)
a1
```

```python
sin_a1 = np.sin(a1)
sin_a1
```

```python
tanh_a1 = np.tanh(a1)
```

```python
sin_a1.shape, tanh_a1.shape
```

```python
diff = sin_a1 - tanh_a1
diff
```

```python
np.max(diff)
np.nanmax(diff)
diff.max()
```

```python
x = np.arange(-2 * np.pi, 2 * np.pi, 0.1)
y = np.sin(x)
y2 = np.tanh(x)
y3 = np.cos(x)
# plt.subplot(311)
plt.plot(x,y, label="sin")
# plt.subplot(312)
plt.scatter(x, y2, color="red", label="tanh")
# plt.subplot(313)
plt.bar(x, y3, color="green", label="cos")
plt.legend()
plt.title("Test")
# plt.show()
```
